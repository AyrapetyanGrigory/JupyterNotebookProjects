```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Импортируем датасет, с которым будем работать:

```python
df = pd.read_csv("hw2.csv")
df
```

Обозначим X и Y:

```python
X = np.array(df["X"])
y = np.array(df["Y"])
```

Импортируем модель линейной регрессии из библиотеки "sklearn":

```python
from sklearn import linear_model
```

Дадим значение переменной для модели линейной регрессии и поменяем размермность наших данных для работы:

```python
reg = linear_model.LinearRegression() 
reg
Z = X.reshape(-1, 1) 
B = y.reshape(-1, 1) 
```

Обучим нашу модель:

```python
reg.fit(Z, B)
```

```python
reg.coef_
```

```python
reg.score(Z, B)
```

Предскажем результаты при помощи модели:

```python
reg.predict(B)
```

```python
reg.predict(Z)
```

```python
reg.intercept_
```
